```python
import numpy as np
```

```python
# Create a numpy array named energy_consumption with energy consumption values in Mwh
energy_consumption = np.array([1200,3400,2900,1800,2500])

# Print the energy consumption array
print("Energy consumption (in Mwh) for different Renewable sources:",energy_consumption)
```

```python
# Calculate the sum of the energy consumption array
sum = np.sum(energy_consumption)

# Print the total energy consumption
print("Total energy consumption (in Mwh) for different Renewable sources:",sum, "Mwh")
```

```python
# Calculate the average of the energy consumption array
average = np.average(energy_consumption)

# Print the average energy consumption
print("Average energy consumption (in Mwh) for different Renewable sources:", average, "Mwh")
```

```python
# Calculate the mean of the energy consumption array
mean = np.mean(energy_consumption)

# Print the mean energy consumption, formatted to two decimal places
print(f"Mean energy consumption (in Mwh) for different Renewable sources: {mean:.2f} Mwh")
```

```python
# Calculate the standard deviation of the energy consumption array
standard = np.std(energy_consumption)

# Print the standard deviation of energy consumption, formatted to two decimal places
print(f"Standard deviation of energy consumption (in Mwh) for different Renewable sources: {standard:.2f} Mwh")
```

```python
# Reshape the energy_consumption array into a 5x1 column vector
reshaped_array = energy_consumption.reshape(5,1)

# Print the reshaped array
print("Reshaped array:",reshaped_array)
```

```python
import pandas as pd
# Create a dictionary named data with energy information
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}
# Create a pandas DataFrame named df from the data dictionary
df=pd.DataFrame(data)
# Print the DataFrame
print(df)
```

```python
# Display the first 5 rows of the DataFrame
df.head()
```

```python
# Display the last 5 rows of the DataFrame
df.tail()
```

```python
# Check for missing values in each column of the DataFrame
df.isnull().sum()
```

```python
# Drop rows with any missing values (NaN) from the DataFrame
cleaned_df = df.dropna()
# Print the DataFrame after dropping missing values
print(cleaned_df)
```

```python
# Fill missing values (NaN) with the mean of the respective columns
df['Energy Consumption (MWh)'] = df['Energy Consumption (MWh)'].fillna(df['Energy Consumption (MWh)'].mean())
df['Cost (Million $)'] = df['Cost (Million $)'].fillna(df['Cost (Million $)'].mean())

df
```

```python
# Calculate 'Cost per MW' by dividing 'Energy Consumption (MWh)' by 'Cost (Million $)'
df['Cost per MW'] = df['Energy Consumption (MWh)'] / df['Cost (Million $)']
df
```

**Data Preprocessing**


```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler ()

# Scale the 'Energy Consumption (MWh)' and 'Cost (Million $)' columns
df[['Energy Consumption (MWh)', 'Cost (Million $)']] = scaler.fit_transform(df[['Energy Consumption (MWh)', 'Cost (Million $)']])
df
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

df[['Energy Consumption (MWh)', 'Cost (Million $)']] = std_scaler.fit_transform(df[['Energy Consumption (MWh)', 'Cost (Million $)']])
df
```
